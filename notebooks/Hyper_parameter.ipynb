```python
from matplotlib import pyplot as plt
from matplotlib.pyplot import subplots_adjust
import pandas as pd

%matplotlib inline
%config InlineBackend.figure_format = 'retina'
```

### 1) Basic Function

```python
# MAE
x = ["30","64","128","256","512"]
y1 = [0.898,0.690,0.572,0.527,0.545]
lower_error1 = [0.0332,0.03192,0.03138,0.02396,0.087225]
upper_error1 = [0.03254,0.03908,0.02722,0.02554,0.188975]

asymmetric_error1 = [lower_error1, upper_error1]

fig, ax0 = plt.subplots(nrows=1, sharex=True,figsize = [6,4])

ax0.errorbar(x, y1, capsize=2,yerr=asymmetric_error1, c = "coral",fmt='o',ls = 'solid',label = "2_reuse")

ax0.set_title('')
ax0.set_xlabel("Number of Basis Functions", size = 14,fontname="Arial")
ax0.set_ylabel("MAE (kcal/mol)", size = 14,fontname="Arial")
plt.ylim(0.3,1)
plt.yticks(fontsize = 12)
ax0.set_xticklabels(x,size = 12,fontname="Arial")
ax0.set_xticklabels(x,size = 12,fontname="Arial");
#plt.savefig("parm_1_basisfunction_mae.pdf",dpi = 300);
```

### 2) Share/Distinct Weights

```python
# MAE
x = ["shared","distinct"]
y1 = [0.5524,0.5029,0.5102,0.526,0.5428]
y2 = [0.4579,0.4522,0.3976,0.4151,0.4113]


fig, ax0 = plt.subplots(nrows=1, sharex=True,figsize = [3,4])

ax0.boxplot([y1,y2],whis="range")

ax0.set_title('')

ax0.set_xlabel("Interaction Block Weights", size = 14,fontname="Arial")
ax0.set_ylabel("MAE (kcal/mol)", size = 14,fontname="Arial")
plt.ylim(0.3,0.6)
plt.yticks(fontsize = 12)
ax0.set_xticklabels(x,size = 12,fontname="Arial")
ax0.set_xticklabels(x,size = 12,fontname="Arial")

subplots_adjust(left = 0.3);
#plt.savefig("parm_2_retainreuseweights_mae.pdf",dpi = 300);
```

### 3) Umax and gap

```python
x = ["3","4","5","6","8","10","20"]
y1 = [0.3491,0.3694,0.3659,0.3745,0.3724,0.3611,0.3819]
y2 = [0.3724,0.3770,0.3886,0.3916,0.4041,0.4052,0.3931]

fig,ax0 = plt.subplots(figsize = [4.5,4])
plt.plot(x,y1,color = "coral", label = "Δμ=0.1", marker = "o")
plt.plot(x,y2, color = "green",label = "Δμ=0.2", marker = "o")
ax0.set_xlabel("$μ_{max} $", size = 14,fontname="Arial")
ax0.set_ylabel("MAE (kcal/mol)", size = 14,fontname="Arial")
#plt.ylim(0.3,0.5)
plt.yticks(fontsize = 12)
ax0.set_xticklabels(x,size = 12,fontname="Arial")
ax0.set_xticklabels(x,size = 12,fontname="Arial")
ax0.legend(loc = "upper left", fontsize = 10)
fig.tight_layout();
#plt.savefig("parm_3_umaxgap_mae.pdf",dpi = 300);
```

### 7) Training Curve

```python
df = pd.read_csv("training_schedule_30_3_20.csv")
df.head()
```

```python
x = df["epoch"]
y_train = df["Cyclic_train"]
y_val = df["Cyclic_val"]
y_train_1 = df["Decay_train"]
y_val_1 = df["Decay_val"]
```

```python
fig,ax0 = plt.subplots(figsize = [8,4])

plt.plot(x, y_train, color = "red",label = "Cyclic_train")
plt.plot(x, y_val, color = "lightcoral", label = "Cyclic_val")
plt.plot(x, y_train_1, color = "green",label = "Decay_train")
plt.plot(x, y_val_1, color = "lightgreen",label = "Decay_val")
plt.xlim(0,400);
plt.ylim(0,10);
ax0.set_xlabel("Number of Epochs", size = 14,fontname="Arial")
ax0.set_ylabel("MAE (kcal/mol)", size = 14,fontname="Arial")
plt.yticks(fontsize = 12)
plt.xticks(fontsize = 12)
ax0.legend(loc = "upper right", fontsize = 14);
#plt.savefig("parm_7_learningcurve_mae.pdf",dpi = 300);
```
