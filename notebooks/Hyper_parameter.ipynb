```python
def draw_graph(x, y, y_1, error, error_1, error_name, save = False, out = None):
    #x = ["1k","5k","10k","15k","25k","50k","75k","100k"] 
    fig = plt.figure(figsize=[6,4.8])
    plt.errorbar(x,y,error, marker='o', color = "coral", label = "DTNN_7ib")
    plt.errorbar(x,y_1,error_1, marker='o',color = "green", label = "TL_QM9$_\mathrm{M}$")
    plt.yticks(size = 12)
    plt.xticks([5000, 15000, 25000, 50000, 75000, 100500], ["5k","15k","25k","50k","75k","100k"],size = 12)
    plt.legend(fontsize = 12)
    #plt.yticks([0.50, 0.55, 0.6,0.65, 0.70],size = 12)
    plt.xlabel("Number of Training Examples", size = 14,fontname="Arial")
    plt.ylabel(error_name + " (kcal/mol)", size = 14,fontname="Arial")
    if save:
        plt.savefig(out,dpi = 300)
```
